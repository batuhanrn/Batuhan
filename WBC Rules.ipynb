```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_excel('HW1_Question2_dataset.xlsx')
df = df[['week','amount']]
df.index = df['week']
df.drop('week',axis = 1, inplace=True)
df = df.groupby(level=0).sum()
df.head()
```

```python
df_52 = df[df.index<=52]

df_52.head()
```

```python
std  = np.std(df_52['amount'])
mean = df_52['amount'].mean()
print('mean: %0.2f'%mean)
print('std: %0.2f'%std)
```

```python
df_test = df
df_test.head()
```

```python
above_3_std =  mean + 3 * std
below_3_std =  mean - 3 * std
above_2_std =  mean + 2 * std
below_2_std =  mean - 2 * std
above_1_std =  mean + 1 * std
below_1_std =  mean - 1 * std
print('above 3 std: %0.2f'%above_3_std)
print('above 2 std: %0.2f'%above_2_std)
print('above 1 std: %0.2f'%above_1_std)
print('below 3 std: %0.2f'%below_3_std)
print('below 2 std: %0.2f'%below_2_std)
print('below 1 std: %0.2f'%below_1_std)
```

```python
# Rule 1 
df_test['Rule-1'] = (df_test['amount'] > above_3_std) | (df_test['amount'] < below_3_std) 
df_test.head()
```

```python
# Rule 2 
df_test['shif-1'] = np.multiply(np.array(df_test['amount'] > mean),1)
df_test['shif-2'] = np.multiply(np.array(df_test.shift(1)['amount'] > mean),1)
df_test['shif-3'] = np.multiply(np.array(df_test.shift(2)['amount'] > mean),1)
df_test['shif-4'] = np.multiply(np.array(df_test.shift(3)['amount'] > mean),1)
df_test['shif-5'] = np.multiply(np.array(df_test.shift(4)['amount'] > mean),1)
df_test['shif-6'] = np.multiply(np.array(df_test.shift(5)['amount'] > mean),1)
df_test['shif-7'] = np.multiply(np.array(df_test.shift(6)['amount'] > mean),1)
df_test['shif-8'] = np.multiply(np.array(df_test.shift(7)['amount'] > mean),1)
df_test['shif-9'] = np.multiply(np.array(df_test.shift(8)['amount'] > mean),1)
# df_test['shif-0..8'] "1" means amount is above the mean, "0" means amount below the mean

df_test['Rule-2'] = (df_test[['shif-1','shif-2','shif-3','shif-4','shif-5','shif-6','shif-7','shif-8','shif-9']]\
                    .sum(axis=1) == 9) | (df_test[['shif-1','shif-2','shif-3','shif-4','shif-5','shif-6'\
                                                 ,'shif-7','shif-8','shif-9']].sum(axis=1) == 0)


df_test.drop(['shif-1','shif-2','shif-3','shif-4','shif-5','shif-6','shif-7','shif-8','shif-9'],axis=1,inplace=True)
```

```python
# Rule 3

df_test['shif-1'] = np.multiply(np.array(df_test['amount'] > df_test.shift(1)['amount']),1)
df_test['shif-2'] = np.multiply(np.array(df_test.shift(1)['amount'] > df_test.shift(2)['amount']),1)
df_test['shif-3'] = np.multiply(np.array(df_test.shift(2)['amount'] > df_test.shift(3)['amount']),1)
df_test['shif-4'] = np.multiply(np.array(df_test.shift(3)['amount'] > df_test.shift(4)['amount']),1)
df_test['shif-5'] = np.multiply(np.array(df_test.shift(4)['amount'] > df_test.shift(5)['amount']),1)
df_test['shif-6'] = np.multiply(np.array(df_test.shift(5)['amount'] > df_test.shift(6)['amount']),1)

df_test['Rule-3'] = (df_test[['shif-1','shif-2','shif-3','shif-4','shif-5','shif-6']].sum(axis=1) == 6) \
                    | (df_test[['shif-1','shif-2','shif-3','shif-4','shif-5','shif-6']].sum(axis=1) == 0)


df_test.drop(['shif-1','shif-2','shif-3','shif-4','shif-5','shif-6'],axis=1,inplace=True)
```

```python
# Rule 4

df_test['shif-1'] = np.multiply(np.array(df_test['amount'] > df_test.shift(1)['amount'])            ,1)
df_test['shif-2'] = np.multiply(np.array(df_test.shift(1)['amount'] > df_test.shift(2)['amount'])   ,10)
df_test['shif-3'] = np.multiply(np.array(df_test.shift(2)['amount'] > df_test.shift(3)['amount'])   ,100)
df_test['shif-4'] = np.multiply(np.array(df_test.shift(3)['amount'] > df_test.shift(4)['amount'])   ,1000)
df_test['shif-5'] = np.multiply(np.array(df_test.shift(4)['amount'] > df_test.shift(5)['amount'])   ,10000)
df_test['shif-6'] = np.multiply(np.array(df_test.shift(5)['amount'] > df_test.shift(6)['amount'])   ,100000)
df_test['shif-7'] = np.multiply(np.array(df_test.shift(6)['amount'] > df_test.shift(7)['amount'])   ,1000000)
df_test['shif-8'] = np.multiply(np.array(df_test.shift(7)['amount'] > df_test.shift(8)['amount'])   ,10000000)
df_test['shif-9'] = np.multiply(np.array(df_test.shift(8)['amount'] > df_test.shift(9)['amount'])   ,100000000)
df_test['shif-10'] = np.multiply(np.array(df_test.shift(9)['amount'] > df_test.shift(10)['amount']) ,1000000000)
df_test['shif-11'] = np.multiply(np.array(df_test.shift(10)['amount'] > df_test.shift(11)['amount']),10000000000)
df_test['shif-12'] = np.multiply(np.array(df_test.shift(11)['amount'] > df_test.shift(12)['amount']),100000000000)
df_test['shif-13'] = np.multiply(np.array(df_test.shift(12)['amount'] > df_test.shift(13)['amount']),1000000000000)
df_test['shif-14'] = np.multiply(np.array(df_test.shift(13)['amount'] > df_test.shift(14)['amount']),10000000000000)

df_test['Rule-4'] = (df_test[['shif-1','shif-2','shif-3','shif-4','shif-5','shif-6','shif-7','shif-8','shif-9',\
                              'shif-10','shif-11','shif-12','shif-13','shif-14']].sum(axis=1) ==   1010101010101) \
                    | (df_test[['shif-1','shif-2','shif-3','shif-4','shif-5','shif-6','shif-7','shif-8',\
                                'shif-10','shif-11','shif-12','shif-13','shif-14']].sum(axis=1) == 101010101010)



df_test.drop(['shif-1','shif-2','shif-3','shif-4','shif-5','shif-6','shif-7','shif-8','shif-9',\
              'shif-10','shif-11','shif-12','shif-13','shif-14'],axis=1,inplace=True)

```

```python
# Rule 5

df_test['b-shif-1'] = np.multiply(np.array(df_test['amount'] < below_2_std)            ,1)
df_test['b-shif-2'] = np.multiply(np.array(df_test.shift(1)['amount'] < below_2_std)   ,1)
df_test['b-shif-3'] = np.multiply(np.array(df_test.shift(2)['amount'] < below_2_std)   ,1)

df_test['a-shif-1'] = np.multiply(np.array(df_test['amount'] > above_2_std)            ,1)
df_test['a-shif-2'] = np.multiply(np.array(df_test.shift(1)['amount'] > above_2_std)   ,1)
df_test['a-shif-3'] = np.multiply(np.array(df_test.shift(2)['amount'] > above_2_std)   ,1)

df_test['Rule-5'] =   (df_test[['b-shif-1','b-shif-2','b-shif-3']].sum(axis=1) >= 2) \
                    | (df_test[['a-shif-1','a-shif-2','a-shif-3']].sum(axis=1) >= 2)



df_test.drop(['b-shif-1','b-shif-2','b-shif-3',\
              'a-shif-1','a-shif-2','a-shif-3'],axis=1,inplace=True)
```

```python
# Rule 6

df_test['b-shif-1'] = np.multiply(np.array(df_test['amount'] < below_1_std)            ,1)
df_test['b-shif-2'] = np.multiply(np.array(df_test.shift(1)['amount'] < below_1_std)   ,1)
df_test['b-shif-3'] = np.multiply(np.array(df_test.shift(2)['amount'] < below_1_std)   ,1)
df_test['b-shif-4'] = np.multiply(np.array(df_test.shift(3)['amount'] < below_1_std)   ,1)
df_test['b-shif-5'] = np.multiply(np.array(df_test.shift(4)['amount'] < below_1_std)   ,1)

df_test['a-shif-1'] = np.multiply(np.array(df_test['amount'] > above_1_std)            ,1)
df_test['a-shif-2'] = np.multiply(np.array(df_test.shift(1)['amount'] > above_1_std)   ,1)
df_test['a-shif-3'] = np.multiply(np.array(df_test.shift(2)['amount'] > above_1_std)   ,1)
df_test['a-shif-4'] = np.multiply(np.array(df_test.shift(3)['amount'] > above_1_std)   ,1)
df_test['a-shif-5'] = np.multiply(np.array(df_test.shift(4)['amount'] > above_1_std)   ,1)

df_test['Rule-6'] =   (df_test[['b-shif-1','b-shif-2','b-shif-3','b-shif-4','b-shif-5']].sum(axis=1) >= 4) \
                    | (df_test[['a-shif-1','a-shif-2','a-shif-3','a-shif-4','a-shif-5']].sum(axis=1) >= 4)


df_test.drop(['b-shif-1','b-shif-2','b-shif-3','b-shif-4','b-shif-5',\
              'a-shif-1','a-shif-2','a-shif-3','a-shif-4','a-shif-5'],axis=1,inplace=True)
```

```python
# Rule 7

df_test['shif-1'] = (np.multiply((np.array(df_test['amount'] < above_1_std)) & \
                                 (np.array(df_test['amount'] > below_1_std)),1))

df_test['shif-2'] = (np.multiply((np.array(df_test.shift(1)['amount'] < above_1_std)) & \
                        (np.array(df_test.shift(1)['amount'] > below_1_std)),1))

df_test['shif-3'] = (np.multiply((np.array(df_test.shift(2)['amount'] < above_1_std)) & \
                        (np.array(df_test.shift(2)['amount'] > below_1_std)),1))

df_test['shif-4'] = (np.multiply((np.array(df_test.shift(3)['amount'] < above_1_std)) & \
                        (np.array(df_test.shift(3)['amount'] > below_1_std)),1))

df_test['shif-5'] = (np.multiply((np.array(df_test.shift(4)['amount'] < above_1_std)) & \
                        (np.array(df_test.shift(4)['amount'] > below_1_std)),1))
                     
df_test['shif-6'] = (np.multiply((np.array(df_test.shift(5)['amount'] < above_1_std)) & \
                        (np.array(df_test.shift(5)['amount'] > below_1_std)),1))

df_test['shif-7'] = (np.multiply((np.array(df_test.shift(6)['amount'] < above_1_std)) & \
                        (np.array(df_test.shift(6)['amount'] > below_1_std)),1))

df_test['shif-8'] = (np.multiply((np.array(df_test.shift(7)['amount'] < above_1_std)) & \
                        (np.array(df_test.shift(7)['amount'] > below_1_std)),1))

df_test['shif-9'] = (np.multiply((np.array(df_test.shift(8)['amount'] < above_1_std)) & \
                        (np.array(df_test.shift(8)['amount'] > below_1_std)),1))

df_test['shif-10'] = (np.multiply((np.array(df_test.shift(9)['amount'] < above_1_std)) & \
                        (np.array(df_test.shift(9)['amount'] > below_1_std)),1))

df_test['shif-11'] = (np.multiply((np.array(df_test.shift(10)['amount'] < above_1_std)) & \
                        (np.array(df_test.shift(10)['amount'] > below_1_std)),1))

df_test['shif-12'] = (np.multiply((np.array(df_test.shift(11)['amount'] < above_1_std)) & \
                        (np.array(df_test.shift(11)['amount'] > below_1_std)),1))

df_test['shif-13'] = (np.multiply((np.array(df_test.shift(12)['amount'] < above_1_std)) & \
                        (np.array(df_test.shift(12)['amount'] > below_1_std)),1))

df_test['shif-14'] = (np.multiply((np.array(df_test.shift(13)['amount'] < above_1_std)) & \
                        (np.array(df_test.shift(13)['amount'] > below_1_std)),1))

df_test['shif-15'] = (np.multiply((np.array(df_test.shift(14)['amount'] < above_1_std)) & \
                        (np.array(df_test.shift(14)['amount'] > below_1_std)),1))



df_test['Rule-7'] = (df_test[['shif-1','shif-2','shif-3','shif-4','shif-5','shif-6','shif-7','shif-8','shif-9',\
                              'shif-10','shif-11','shif-12','shif-13','shif-14','shif-15']]
                    .sum(axis=1) == 15) 
df_test.drop(['shif-1','shif-2','shif-3','shif-4','shif-5','shif-6','shif-7','shif-8','shif-9',\
              'shif-10','shif-11','shif-12','shif-13','shif-14','shif-15'],axis=1,inplace=True)
```

```python
#Rule 8

df_test['shif-1'] = (np.multiply((np.array(df_test['amount'] > above_1_std)) | \
                        (np.array(df_test['amount'] < below_1_std)),1))

df_test['shif-2'] = (np.multiply((np.array(df_test.shift(1)['amount'] > above_1_std)) | \
                        (np.array(df_test.shift(1)['amount'] < below_1_std)),1))

df_test['shif-3'] = (np.multiply((np.array(df_test.shift(2)['amount'] > above_1_std)) | \
                        (np.array(df_test.shift(2)['amount'] < below_1_std)),1))

df_test['shif-4'] = (np.multiply((np.array(df_test.shift(3)['amount'] > above_1_std)) | \
                        (np.array(df_test.shift(3)['amount'] < below_1_std)),1))

df_test['shif-5'] = (np.multiply((np.array(df_test.shift(4)['amount'] > above_1_std)) | \
                        (np.array(df_test.shift(4)['amount'] < below_1_std)),1))

df_test['shif-6'] = (np.multiply((np.array(df_test.shift(5)['amount'] > above_1_std)) | \
                        (np.array(df_test.shift(5)['amount'] < below_1_std)),1))

df_test['shif-7'] = (np.multiply((np.array(df_test.shift(6)['amount'] > above_1_std)) | \
                        (np.array(df_test.shift(6)['amount'] < below_1_std)),1))

df_test['shif-8'] = (np.multiply((np.array(df_test.shift(7)['amount'] > above_1_std)) | \
                        (np.array(df_test.shift(7)['amount'] < below_1_std)),1))

df_test['Checker'] = (np.multiply((
                        (np.array(df_test['amount'] > above_1_std)) & \
                        (np.array(df_test.shift(1)['amount'] > above_1_std)) & \
                        (np.array(df_test.shift(2)['amount'] > above_1_std)) & \
                        (np.array(df_test.shift(3)['amount'] > above_1_std)) & \
                        (np.array(df_test.shift(4)['amount'] > above_1_std)) & \
                        (np.array(df_test.shift(5)['amount'] > above_1_std)) & \
                        (np.array(df_test.shift(6)['amount'] > above_1_std)) & \
                        (np.array(df_test.shift(7)['amount'] > above_1_std)))|
                                    (
                        (np.array(df_test['amount'] < below_1_std)) & \
                        (np.array(df_test.shift(1)['amount'] < below_1_std)) & \
                        (np.array(df_test.shift(2)['amount'] < below_1_std)) & \
                        (np.array(df_test.shift(3)['amount'] < below_1_std)) & \
                        (np.array(df_test.shift(4)['amount'] < below_1_std)) & \
                        (np.array(df_test.shift(5)['amount'] < below_1_std)) & \
                        (np.array(df_test.shift(6)['amount'] < below_1_std)) & \
                        (np.array(df_test.shift(7)['amount'] < below_1_std)))
                                                                             ,1))

df_test['Rule-8'] = (df_test[['shif-1','shif-2','shif-3','shif-4','shif-5','shif-6','shif-7','shif-8']]\
                    .sum(axis=1) == 8)  & \
                    (df_test[['shif-1','shif-2','shif-3','shif-4','shif-5','shif-6','shif-7','shif-8','Checker']]\
                    .sum(axis=1) != 9)

df_test.drop(['shif-1','shif-2','shif-3','shif-4','shif-5','shif-6','shif-7','shif-8','Checker'],axis=1,inplace=True)
```
