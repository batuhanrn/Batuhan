#### .../5 1a. Define the function `P(a,x)` that returns $$\sum_{i=0}^n a_{i}x^{(n-i)}$$
#### Here, `a` is a list which contains `n+1` floating numbers inside; `x` is a floating number.
#### Hint: $a_{k}\equiv $ `a[k]` where k is an integer.
In this question, you are **not allowed** to import and/or use any library. You are allowed to use Python's default classes/functions unless you import anything.

````python
def P(a,x):
    answer_1 = 0 
    n = len(a)-1
    
    for i in range(len(a)):
        answer_1 += a[i]*(x**(n-i))
        
    return answer_1
    
````

## .../15 Q4. Object Oriented Programming
#### .../10 4.a Define the class `TextProcess()`. Define the initial function with the proper self variables. Put all the functions you defined in Q3
#### `textsplitter, bracketremover, textlist_converter, unique_creator, mapping, list_to_vec, joining` inside the class.
#### Define the function `converter(text)` inside the class which takes a string inside, runs all the functions by order and
#### returns mapped version of the string `text`.
In this question, you are **not allowed** to import and/or use any library. You are allowed to use Python's default classes/functions unless you import anything.

````python
import re
class TextProcess:

    def _init_(self):
        pass
    
    def textsplitter(self, text):
        text = text.lower()
        text_list = text.split()
        return text_list
    

    def bracketremover(self, text):
        pattern = r'\[[^()]*\]'
        txt = re.sub(pattern, '', text)
        return txt
    
    def textlist_converter(self, textlist):
        textlist_cleaned=[]
        for i in textlist:
            textlist_cleaned.append(bracketremover(i))
        return textlist_cleaned
    
    def unique_creator(self, textlist_cleaned):
        textlist_unique = list(set(textlist_cleaned))
        return textlist_unique
    
    def mapping(self, textlist_unique):
        numbers = []
        for i in range(len(textlist_unique)):
            numbers.append(i)
        result = dict(zip(textlist_unique, numbers))
        return result
    
    def list_to_vec(self, textlist_cleaned,textlist_map):
        textlist_vec = []
        for i in textlist_cleaned:
            textlist_vec.append(textlist_map[i])
        return textlist_vec
    
    def joinning(self, alist):
        alist = map(str, alist)
        result = " ".join(alist)
        return result
    
    def converter(self, text):
        self.cleaned = []
        self.textlist = self.textsplitter(text)
        for i in self.textlist:
            self.cleaned.append(self.bracketremover(i))
        self.list_cleaned = self.textlist_converter(self.cleaned)
        self.list_unique = self.unique_creator(self.list_cleaned)
        self.list_map = self.mapping(self.list_unique)
        self.list_vec = self.list_to_vec(self.list_cleaned,self.list_map)
        string_all = self.joinning(self.list_vec)
        return string_all
````

#### Do not change this cell and test your code here. It should give the obtained output. (Orders could be different, no problem!)

````python
text_test = text
tp = TextProcess()
converted = tp.converter(text_test)
converted[:40]
````

#### .../15 5c. Define the function `ball_prob(ball_box_dict, ball_draw_dict, n_draw, n_sim, replacement)` that simulates and returns the probability of drawing balls from a box.

Calculate the probability of drawing exactly 1 white ball and 2 black balls by writing a simulation. Run the simulation 10000 times.


````python
import collections
def ball_prob(ball_box_dict, ball_draw_dict, n_draw, n_sim=10000, replacement=False):

    resulut_list = []
    
    if replacement==False:
        for trial in range(n_sim):
            choice = []

            ball_box_dict_1 = ball_box_dict.copy()
            for i in range(n_draw):
                probs = []
                for n_balls in ball_box_dict_1.values():
                    probs.append(n_balls/sum(ball_box_dict_1.values()))

                a = np.random.choice(list(ball_box_dict_1.keys()), size=(1,1), p=probs)
                b = list(a[0])
                choice.append(b[0])
                ball_box_dict_1[b[0]] = ball_box_dict_1[b[0]]-1

            result = dict(collections.Counter(choice))

            test=0
            for key, values in result.items():
                if ball_draw_dict.get(key) == values:
                    test += 1
            if test == len(ball_draw_dict.items()):
                resulut_list.append(1)    
            else:
                resulut_list.append(0)
       
    elif replacement==True:
        for trial in range(n_sim):
            choice = []

            ball_box_dict_1 = ball_box_dict.copy()
            for i in range(n_draw):
                probs = []
                for n_balls in ball_box_dict_1.values():
                    probs.append(n_balls/sum(ball_box_dict_1.values()))

                a = np.random.choice(list(ball_box_dict_1.keys()), size=(1,1), p=probs)
                b = list(a[0])
                choice.append(b[0])

            result = dict(collections.Counter(choice))
            test=0
            for key, values in result.items():
                if ball_draw_dict.get(key) == values:
                    test += 1
            if test == len(ball_draw_dict.items()):
                resulut_list.append(1)    
            else:
                resulut_list.append(0)
                
                
    prob = sum(resulut_list)/len(resulut_list)
    return round(prob,2)
    #your code here
````
